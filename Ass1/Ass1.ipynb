```python
import matplotlib.pyplot as plt
import numpy as np

#defining the Sellmeier coefficients for each material
#Format: Material = [[a1,a2,a3],[b1,b2,b3]]

FuSi = [[0.6961663,0.4079426,0.8974794],[0.0684043,0.1162414,9.896161]]
CaF2 = [[0.5675888,0.4710914,3.8484723],[0.050263605,0.1003909,34.649040]]
SF11 = [[1.7848403,0.311168974,1.17490871],[0.11664845,0.24818551,11.041861754]]

#defining the function for refractive index, group velocity & group vel. dispersion

def ref_index(mat,l): #wavelengths to be input in microns, not metres
    a = 1
    for i in range(3):
        a += (mat[0][i])*(l**2)/(l**2 - (mat[1][i])**2)
    return np.sqrt(a) #Sellmeier formula

def d1_ri(mat,l,h=0.001): #defining the derivative of ref. index for further use
    return (ref_index(mat,l+h)-ref_index(mat,l))/h

c = 3e8 #speed of light in m/s    

def grp_vel(mat,l):
    return c/(ref_index(mat,l)-l*d1_ri(mat,l))

def d2_ri(mat,l,h=0.001): #defining the second derivative of ref. index for further use
    return (d1_ri(mat,l+h)-d1_ri(mat,l)/h)

def gvd(mat,l):
    return ((l**3)/(2*(np.pi)*(c**2)))*d2_ri(mat,l)/(1e6)
```

```python
#plotting the refractive indices

X = np.arange(0.2,1,0.01) #200 nm to 1000 nm (visible light plus a little more)

#plot for fused silica

Y1 = ref_index(FuSi,X)

plt.plot(X,Y1)
plt.xlabel("Wavelength (micron)")
plt.ylabel("Refractive Index")
plt.title("Refractive Index of Fused Silica")

plt.show()
```

```python
#plot for CaF2

Y2 = ref_index(CaF2,X)

plt.plot(X,Y2)
plt.xlabel("Wavelength (micron)")
plt.ylabel("Refractive Index")
plt.title("Refractive Index of CaF2")

plt.show()
```

```python
#plot for SF11

X3 = np.arange(0.3,3,0.01) #300 nm to 3000 nm
Y3 = ref_index(SF11,X3)

plt.plot(X3,Y3)
plt.xlabel("Wavelength (micron)")
plt.ylabel("Refractive Index")
plt.title("Refractive Index of SF11")

plt.show()
```

```python
#plotting the group velocities

#plot for Fused Silica

YY1 = grp_vel(FuSi,X)

plt.plot(X,YY1)
plt.xlabel("Wavelength (micron)")
plt.ylabel("Group Velocity (m/s)")
plt.title("Group Velocity of Fused Silica")

plt.show()
```

```python
#plot for CaF2

YY2 = grp_vel(CaF2,X)

plt.plot(X,YY2)
plt.xlabel("Wavelength (micron)")
plt.ylabel("Group Velocity (m/s)")
plt.title("Group Velocity of CaF2")

plt.show()
```

```python
#plot for SF11

YY3 = grp_vel(SF11,X3)

plt.plot(X3,YY3)
plt.xlabel("Wavelength (micron)")
plt.ylabel("Group Velocity (m/s)")
plt.title("Group Velocity of SF11")

plt.show()
```

```python
#plotting the group velocity dispersion

#plot for fused silica

YYY1 = gvd(FuSi,X)

plt.plot(X,YYY1)
plt.xlabel("Wavelength (micron)")
plt.ylabel("GVD (s^2/m)")
plt.title("GVD of Fused Silica")

plt.show()
```

```python
#plot for CaF2

YYY2 = gvd(CaF2,X)

plt.plot(X,YYY2)
plt.xlabel("Wavelength (micron)")
plt.ylabel("GVD (s^2/m)")
plt.title("GVD of CaF2")

plt.show()
```

```python
#plot for fused silica

YYY3 = gvd(SF11,X3)

plt.plot(X3,YYY3)
plt.xlabel("Wavelength (micron)")
plt.ylabel("GVD (s^2/m)")
plt.title("GVD of SF11")

plt.show()
```
